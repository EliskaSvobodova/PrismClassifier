```python
import pandas as pd
```

```python
airline_train = pd.read_csv("data/airline/train.csv")
airline_test = pd.read_csv("data/airline/test.csv")
airline = pd.concat([airline_train, airline_test], ignore_index=True).drop(['Unnamed: 0', 'id'], axis=1)
airline = airline.sample(3100, random_state=42).dropna()
airline.to_csv('data/airline/data.csv', index=False)
airline # satisfaction
```

```python
col = ["Industrial Risk", "Management Risk", "Financial Flexibility", "Credibility", "Competitiveness", "Operating Risk", "Bankruptcy"]
bank = pd.read_csv("data/bankruptcy/Qualitative_Bankruptcy.data.txt", names=col)
bank.to_csv("data/bankruptcy/data.csv", index=False)
bank
```

```python
cancer = pd.read_csv("data/cancer/breast-cancer-wisconsin.data", names=['id', 'Clump Thickness', 'Uniformity of Cell Size', 'Uniformity of Cell Shape', 'Marginal Adhesion', 'Single Epithelial Cell Size', 'Bare Nuclei', 'Bland Chromatin', 'Normal Nucleoli', 'Mitoses', 'Class']).drop('id', axis=1)
cancer['Class'] = cancer['Class'].replace({2: "benign", 4: "malignant"})
cancer.to_csv("data/cancer/data.csv", index=False)
cancer # Class
```
